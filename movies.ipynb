```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
df = pd.read_csv(f'D:\PythonProjects\movies.csv')
df.head()
```

```python
# Data types
df.dtypes
```

```python
# Find missing values
for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing))
```

```python
# Total null values
sum(df.isnull().values.ravel())
```

```python
# Total rows with null values
sum([True for idx,row in df.iterrows() if any(row.isnull())])
```

```python
# For now we will replace these null values with 0
df = df.fillna(0)
df.head()
```

```python
# Find missing values
for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing))
```

```python
# Change data types to remove decimals
df['votes'] = df['votes'].astype('int64')
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
df.head()
```

```python
# New column 'release date' from released as datetime
df['release date'] = pd.to_datetime(df['released'].str.split(' \(').str[0])
# Correcting year (for now na values are filled with 0 to remove decimal from year)
df['correctyear'] = df['release date'].dt.year
df['correctyear'] = df['correctyear'].fillna(0)
df['correctyear'] = df['correctyear'].astype('int64')
df.head()
```

```python
# Show only 4 rows
pd.set_option('display.max_rows', 8)
```

```python
# Sortiing by gross revenue
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop duplicates
df['name'].drop_duplicates().sort_values(ascending=False)
```

```python
new_df = df.loc[~(df==0).all(axis=1)]
new_df= df[df['company'] != 0]
```

```python
# Gross revenue of companies in specific years
new_df.groupby(['company', 'year'])[["gross"]].sum()
```

```python
# Top 5 compaies by gross revenue
gross_revenue = df.groupby('company')[["gross"]].sum()
gross_revenue_sorted = gross_revenue.sort_values('gross', ascending = False)[:5]
gross_revenue_sorted = gross_revenue_sorted['gross'].astype('int64') 
gross_revenue_sorted
```

```python
correlation_matrix = df.corr(numeric_only=(True))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Attributes')
plt.ylabel('Attributes')
plt.show()
```

```python
# Assigning numerical values to objects
df_numbered = df

for col_name in df_numbered.columns:
    if(df_numbered[col_name].dtype == 'object'):
        df_numbered[col_name]= df_numbered[col_name].astype('category')
        df_numbered[col_name] = df_numbered[col_name].cat.codes
        
df_numbered
```

```python
df_numbered_corr = df_numbered.corr(method='pearson', numeric_only=True)
df_numbered_corr
```

```python
sns.heatmap(df_numbered_corr, annot = True, annot_kws={'size':5})
plt.title("Correlation matrix for Movies")
plt.xlabel("Movie Attributes")
plt.ylabel("Movie Attributes")
plt.show()
```

```python
# Assigning a random numeric value for each unique categorical value using factorize

df_numerized = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
df_numerized.head()
```

```python
# Correlation matrix for the numerized attributes

sns.heatmap(df_numerized, annot = True, cmap="YlGnBu", linewidths=0.0, annot_kws={'size':5})
plt.title("Correlation matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")

plt.show()
```

```python
corr_pairs = df_numerized.unstack()
print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
print(sorted_pairs)
```

```python
# Highly correlated be greeater than 5.
highly_correlated = sorted_pairs[abs(sorted_pairs) > 0.5]
print(highly_correlated)
```
